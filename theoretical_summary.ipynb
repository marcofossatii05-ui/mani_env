# **Approssimazione ai volumi finiti**
Consideriamo la legge di conservazione
</br></br>
\begin{equation*}
\frac{\partial c}{\partial t} + \frac{\partial f(c)}{\partial x} = 0 \quad \text{in}\ \Omega\times[0,T)
\end{equation*}
</br>
dove l'incognita $c(x,t)$ è una quantità che viene conservata nel tempo e $f:\mathbb{R}\longrightarrow\mathbb{R}$ è detta funzione di flusso. Imponiamo la seguente **condizione iniziale**
</br></br>
\begin{equation*}
	c(x,t=0) = c_0(x) \quad \text{in}\ \Omega.
\end{equation*}
</br>
Nel caso monodimensionale, in cui il dominio spaziale è un intervallo $\Omega=[a,b]$, diremo che l'estremo $a$ è

* un bordo di *inflow* se $f'(c(a,t))>0$,
* un bordo di *outflow* se $f'(c(a,t))<0$,

diremo che l'estremo $b$ è

* un bordo di *inflow* se $f'(c(b,t))<0$,
* un bordo di *outflow* se $f'(c(b,t))>0$.

mentre un estremo dell'intervallo $[a,b]$ è detto di *noflow* se in quel punto $f'(c)=0$.


Imponiamo quindi una **condizione al bordo** sull'estremo di inflow $\partial\Omega_{in}$:
</br></br>
$$
	c = \bar{c} \quad \text{in}\ \partial\Omega_{in}.
$$

## **Discretizzazione in tempo**
Suddividiamo il dominio temporale in $N_T$ sottointervalli di uguale ampiezza $\Delta t=\frac{T}{N_T}$ e consideriamo la griglia temporale data dai loro estremi:
</br></br>
$$
	0 = t_0 < t_1 < \dots < t_n < \dots < T_{N_t} = T
$$
dove $t_n = n\Delta t$, con $n=0,\dots,N_T$.
</br></br>
Applicando il metodo di Eulero esplicito otteniamo il sistema semi-discretizzato in tempo:
</br>
</br>
$$
	\frac{c(x,t_{n+1})-c(x,t_n)}{\Delta t} + \frac{\partial f(c(x,t_n))}{\partial x} = 0 \quad \forall x\in\Omega, \ n=0,\dots,N_T-1. \quad\quad\quad (1.1)
$$

## **Discretizzazione in spazio**
Suddividiamo ora il dominio spaziale $\Omega$ in $N$ sottointervalli, di uguale ampiezza $h$. Chiamiamo $K_i$ la cella $i$-esima di questa griglia, per $i=1,\dots,N$ e siano

* $x_i$ il punto medio di $K_i$
* $x_{i-\frac{1}{2}}=x_i - \frac{h}{2}$, estremo di sinistra (punto di bordo tra $K_{i-1}$ e $K_i$).
* $x_{i+\frac{1}{2}}= x_i +\frac{h}{2}$, estremo di destra (punto di bordo tra $K_{i}$ e $K_{i+1}$).


Vogliamo costruire una soluzione numerica che in ogni istante temporale sia costante a tratti sulle celle:
</br></br>
\begin{equation*}
c_i^n := \frac{1}{h}\int_{x_{i-\frac{1}{2}}}^{x_{i+\frac{1}{2}}} c(x,t_n) dx, 
\end{equation*}
La soluzione numerica sarà quindi costante a tratti sulle celle e data da 
\begin{equation*}
c_h(x,t) = c_i^n \quad \text{per} \ x\in\left(x_{i-\frac{1}{2}},x_{i+\frac{1}{2}}\right),\ t\in[t_n,t_{n+1}), \ n\geq1.
\end{equation*}
</br>
Il valore $c^0$ è l'approssimazione della condizione iniziale calcolata nello stesso modo
</br></br>
\begin{equation*}
	c_i^0 := \int_{x_{i-\frac{1}{2}}}^{x_{i+\frac{1}{2}}} c_0(x)dx \quad \forall K_i
\end{equation*}
</br>
Integrando (in spazio) su ogni cella l'equazione (1.1) si ottiene quindi
</br></br>
\begin{equation*}
	h \frac{c_i^{n+1}-c_i^n}{\Delta t} + f\left(c\left(x_{i+\frac{1}{2}},t_n\right)\right) -f\left(c\left(x_{i-\frac{1}{2}},t_n\right)\right) = 0 \quad \quad \quad (1.2)
\end{equation*}
</br>


## **Scelta del flusso numerico**
Poiché abbiamo costruito una soluzione costante a tratti sulle celle, questa non è definita sui loro estremi. Approssimiamo quindi il flusso nei nodi della griglia attraverso un ''flusso numerico'' $F$ che descriva lo scambio della quantità $c$ tra celle vicine:
\begin{equation*}
	\begin{cases}
		f(c(x_{i-\frac{1}{2}},t_n)) \approx 	F_{{i-\frac{1}{2}}}^n(c^n_{i-1},c^n_i), \\
		f(c(x_{i+\frac{1}{2}},t_n)) \approx F_{{i+\frac{1}{2}}}^n(c^n_{i},c^n_{i+1})
	\end{cases}
\end{equation*}
e che soddisfi le seguenti ipotesi:
* $F$ localmente Lispschitziano in entrambe le variabili:
\begin{equation*}
	\begin{aligned}
  &\exists k_1>0 \:\ |F(a,b)-F(c,b)| \leq k_1 |a-c|,\\
  &\exists k_2>0 \:\ |F(a,b)-F(a,c)| \leq k_2 |b-c|;
  \end{aligned}
	\end{equation*}
* F è consistente: $F(a,a) = f(a)$;
* F è non-descrescente rispetto ad $a$ e non-crescente rispetto a $b$:
\begin{equation*}
	\frac{\partial F(a,b)}{\partial a} \geq 0, \qquad \frac{\partial F(a,b)}{\partial b} \leq 0.
\end{equation*}
Sostituendo tale approssimazione nell'equazione (1.2) otteniamo uno schema conservativo:
\begin{equation*}
	h \frac{c_i^{n+1}-c_i^n}{\Delta t} + F^n_{i+\frac{1}{2}}(c^n_i,c^n_{i+1}) - F^n_{i-\frac{1}{2}}(c^n_{i-1},c^n_i)= 0, \qquad i=1,\dots,N,\ n =0,\dots,N_T-1.
\end{equation*}

Se il flusso $f(c)$ è monotono, cioè $f'(c)>0\, \forall c\in[c_m,c_M]$ oppure $f'(c)<0\, \forall c\in[c_m,c_M]$, con $c_m = \min_{x\in\Omega}c_0(x)$ e $c_M = \max_{x\in\Omega}c_0(x)$, allora può essere approssimato con il **flusso upwind**, definito come segue:
\begin{equation*}
	F_{i+\frac{1}{2}}^{\text{UP}} (c_i,c_{i+1}) :=
	\begin{cases}
		f(c_i), & \text{se}\ f'(c)\geq 0 \ \text{in}\ [c_i,c_{i+1}], \\
		f(c_{i+1}), & \text{se}\ f'(c)\leq 0 \ \text{in}\ [c_i,c_{i+1}].
	\end{cases}
\end{equation*}
Altrimenti possiamo utilizzare il **flusso Godunov**, basato sul metodo delle caratteristiche. Infatti quando il flusso non è monotono possono verificarsi fenomeni di shock o rarefazione, ed è possibile trovare il valore di $c^* = c\left(x_{i+\frac{1}{2}},t\right)$, $t\in[t^n,t^{n+1}]$, tramite il metodo delle caratteristiche, e quindi calcolare $f(c^*)$ all'interfaccia tra due celle vicine. Il flusso Godunov è definito come segue:

</br></br>
$$
	F_{i+\frac{1}{2}}^{\text{G}} (c_i,c_{i+1}) :=
	\begin{cases}
		\min_{\xi\in[c_i,c_{i+1}]} f(\xi), & \text{se}\ c_i\leq c_{i+1} \\
		\max_{\xi\in[c_i,c_{i+1}]} f(\xi), & \text{se}\ c_i\geq c_{i+1}.
	\end{cases}
$$
</br>

## **Stabilità**
Il metodo dei Volumi Finiti è stabile solo per alcune scelte del passo spaziale $h$ e del passo temporale $\Delta t$, che devono soddisfare la condizione **CFL** (Courant-Friedrichs-Lewy):
\begin{equation*}
	\frac{\Delta t L}{h}\leq 1
\end{equation*}
con $L$ tale che $|f'(c)|\leq L$ $\forall c\in[c_m,c_M]$.

Nel caso di flusso lineare, ovvero $f(c)=vc$ con $v\in\mathbb{R}$, la condizione di stabilità può essere riscritta come
\begin{equation}\nonumber
	\frac{\Delta t |v|}{h}\leq 1.
\end{equation}

Tale condizione assicura che la soluzione in ogni cella ad ogni istante di tempo sia una combinazione lineare convessa dei valori assunti nelle celle vicine. Quando non viene rispettata si osservano nella soluzione numerica oscillazioni spurie non fisiche.
